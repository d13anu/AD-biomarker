## Explore data

```
print(exp.value_counts(['RID']))
print(exp.value_counts(['VISCODE2']))
```

```
# Only baseline data
print("original shape",exp.shape)
exp_bl = exp[exp['VISCODE2']=='bl']
print("only baseline", exp_bl.shape)

# Get ChemID with most missing vales
print("\n\nmeta with missing vals:\n", exp_bl.isnull().sum().sort_values(ascending=False))

# get samples with most missing values
print("\n\nsamples with missing vals:\n",exp_bl.isnull().sum(axis=1).sort_values(ascending=False))
```

```
# drop missing vales columns and rows
# Compounds with more than 20% missing values
exp_bl = exp_bl.dropna(thresh=exp_bl.shape[0] * 0.8, axis=1)

# samples with more than 10% values
exp_bl = exp_bl.dropna(thresh=exp_bl.shape[1] * 0.9, axis=0)

print("\n\nsamples with missing vals:\n",exp_bl.isnull().sum(axis=1).sort_values(ascending=False))
```

```
exp_bl
```

## Normalization and Proprocessing

Reference: ADNI_CruchagaLab_CSF_Metabolon_Metabolomic_Data_20230817.pdf

*Located in: ../Articles*

```
## Transforming, normalizing and standardizing the data
comp_cols = [col for col in exp_bl.columns if col.startswith('X')]

# Apply natural log transformation
exp_bl[comp_cols] = exp_bl[comp_cols].apply(np.log)

# Drop non-essential columns
exp_data_final = exp_bl.drop(['EXAMDATE', 'GUSPECID',
                                'VISCODE2', 'Metabolon_Barcode'], axis=1)
exp_data_final
```

```
# Replacing the remaining NaN expression values with the average of the column
exp_data_final = exp_data_final.fillna(exp_data_final.drop(['RID'], axis=1).mean())

exp_data_final
```

```
exp_data_final.to_csv('SourceData/meta_exp_data.csv', index=False)
```
